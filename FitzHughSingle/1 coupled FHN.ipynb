1 coupled Fitz Hugh Nagumo for coupling investigation

Initial parameters

```python
"""Alpha values"""
a1 = -0.1
"""Epsilon values"""
e1 = 0.01
"""Beta values"""
b1 = 2
```

Solving the 1 coupled FHN

```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.legend import Legend

def fhn(x, t):

    V1 = x[0]
    R1 = x[1]

    dV1dt = (V1 * (a1 - V1) * (V1 - 1) - R1)
    dR1dt = (e1 * (V1 - b1 * R1))

    return [dV1dt,dR1dt]

x0 = [0.1,0] # initial conditions
t = np.linspace(0,500,1000) # time
x = odeint(fhn,x0,t)

V1 = x[:,0]
R1 = x[:,1]
```

Plotting the 3 coupled FHN

```python
fig, (ax0) = plt.subplots(figsize=(10,10))
plt.suptitle('Plot of the FHN model')


"""Graph for the SA Subplot """

#ax0.title.set_text('SA')
ax0.plot(t,V1, label='V1', color='b')  # V1 label
ax0.plot(t,R1, label='V1', color='r')  # V1 label

ax0.plot([ ],[ ],' ',label=r'$\alpha_1$ ={}'.format(a1))   # alpha1
ax0.plot([ ],[ ],' ',label=r'$\beta_1$ ={}'.format(b1))     # beta1
ax0.plot([ ],[ ],' ',label=r'$\epsilon_1$ ={}'.format(e1))     # epsilon1
ax0.legend(bbox_to_anchor=(0.85,0.5), loc="center left", borderaxespad=9)

plt.legend()

ax0.legend(bbox_to_anchor=(0.85,0.5), loc="center left", borderaxespad=9.5)

ax0.grid()
ax0.set_yticks([-0.5, -0.5, 0, 0.5, 0.5, 1])
fig.supylabel('Voltage (mV)')
fig.supxlabel('Time (ms)')
fig.savefig('FHN.jpeg', bbox_inches="tight")
plt.show()

```
