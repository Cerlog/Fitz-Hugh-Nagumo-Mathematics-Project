1 coupled Fitz Hugh Nagumo for coupling investigation

Initial parameters

```python
"""Scaling constants k1"""
k1 = 125

"""Alpha values"""
a1 = -0.1


"""Epsilon values"""
e1 = 0.01


"""Beta values"""
b1 = 2


"""Coupling constants f"""
f1 = 1


"""Coupling constants g"""
g1 = 0


"""Coupling constants h"""
h1 = 0
```

Solving the 1 coupled FHN

```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.legend import Legend


def fhn(x, t):
    # constants

    V1 = x[0]
    W1 = x[1]

    dV1dt = k1*(V1 * (a1 - V1) * (V1 - 1) - W1)
    dW1dt = k1*(e1 * (V1 - b1 * W1))

    return [dV1dt,dW1dt]


x0 = [0.1,0] # initial conditions
t = np.linspace(0,5,1000) # time
x = odeint(fhn,x0,t)

V1 = x[:,0]
W1 = x[:,1]
```
