```python
#all imports
import numpy as np
import matplotlib.pyplot as plt
import cmath
%matplotlib inline
```

```python
#fixed constant Beta
beta = 2


list = np.linspace(-5, 5 , num=1000) # range from -5 to 5, with 200 points


def quadratic1(list): # - determinant
    result2 = [] # list for saving all the points

    for alpha in list: # for alpha in a list
        a = -1
        b = 1+alpha
        c = -(alpha+(1/beta))
        d = (b**2) - (4*a*c)
        result = ((-b-cmath.sqrt(d))/(2*a))
        if result.imag: # if the result is imaginary
            result2.append(0) # zero is added to the list
        else:
            result2.append(result) # else the result is added
    return result2


x = quadratic1(list) # passing list of points to the function
y = list
```
