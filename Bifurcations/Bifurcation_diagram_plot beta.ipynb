```python
alpha = -0.1
```

```python
list = np.linspace(-5, 5 , num=200)
def quadratic1(list):
    result2 = []

    for beta in list:
        a = -1
        b = 1+alpha
        c = -(alpha+(1/beta))
        d = (b**2) - (4*a*c)
        result = ((-b-cmath.sqrt(d))/(2*a))
        if result.imag:
            result2.append(0)
        else:
            result2.append(result)
    return result2
```
