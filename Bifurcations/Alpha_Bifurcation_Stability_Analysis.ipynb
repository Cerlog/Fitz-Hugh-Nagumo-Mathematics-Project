```python
#all imports
import numpy as np
import matplotlib.pyplot as plt
import cmath
import math
%matplotlib inline
```

This program itterates through the values of $\alpha$ with fixed $\beta = 2$ and $\epsilon = 0.01$.

Ranges obtained in the investigations in the project are plugged into the equations bellow, the results are evaluated based on the conditions and depending on the stability, the values of V and Alpha gets sorted into the lists and afterwards plotted on a scatter plot seen in the report.



$ V = 0$ \
$ V = ((\alpha+1) + \sqrt{(\alpha+1)^2-4(\alpha+\frac{1}{\beta})})$ \
$ V = ((\alpha+1) - \sqrt{(\alpha+1)^2-4(\alpha+\frac{1}{\beta})})$





```python
listOfAlphas = np.linspace(-10, -0.4142135623730, num = 1000) # generate Alpha values
listOfAlphas[listOfAlphas == -0.4142135623730] = np.nan # exclude the bifurcation point

# fixed parameters
beta = 2
epsilon = 0.01


# list storing V values
V_equationList = []
V_positiveList = []
V0_minus_list = []


#Lists storing the alpha values
AlphaIFSaddle = []
AlphaIFStable = []
AlphaIFUnstable = []

AlphaELIFSaddle = []
AlphaELIFStable = []
AlphaELIFUnstable = []
AlphaELSEUnstableSpiral = []
AlphaELSEStableSpiral = []
ifSaddle = []
ifStable = []
ifUnstable = []
elifSaddle = []
elifStable = []
elifUnstable = []
elseUnstableSpiral = []
elseStableSpiral = []

for alpha in listOfAlphas:
    V_eq_plus = 0.5*((alpha+1)+(cmath.sqrt((alpha+1)**2-4*(alpha+1/beta))))
    V_equationList.append(V_eq_plus)
    T = 2*(alpha + 1) * V_eq_plus - alpha - 3*V_eq_plus**2 - epsilon * beta
    D = (2*(alpha +1 ) * V_eq_plus - alpha - 3*V_eq_plus**2 ) * (- epsilon*beta) + epsilon
    Eigen_1_plus = 0.5*(T + cmath.sqrt(T**2 - 4*D))
    Eigen_2_m =0.5*(T - cmath.sqrt(T**2 - 4*D))
    V_positiveList.append(Eigen_1_plus)
    V0_minus_list.append(Eigen_2_m)


    #T = (2*v*(alpha+1) - alpha - 3*v**2) - (epsilon * beta)
    #D = (2*v*(alpha+1) - alpha - 3*v**2)*(-beta*epsilon)-(-1)*epsilon
    T_sq_minus_D = T**2 - 4*D
    sqrt_T_sq_minus_D = cmath.sqrt(T ** 2 - 4 * D)

    if T_sq_minus_D > 0:
        if (Eigen_1_plus > 0 and Eigen_2_m < 0) or (Eigen_2_m > 0 and Eigen_1_plus < 0):
            ifSaddle.append(V_eq_plus)
            AlphaIFSaddle.append(alpha)
        elif (Eigen_1_plus > Eigen_2_m) > 0 or (Eigen_2_m > Eigen_1_plus ) > 0:
            ifStable.append(V_eq_plus)
            AlphaIFStable.append(alpha)
        else: #(Eigen_1_plus and Eigen_2_m < 0) or (Eigen_2_m and Eigen_1_plus <0 ):
            ifUnstable.append(V_eq_plus)
            AlphaIFUnstable.append(alpha)

    elif T_sq_minus_D == 0:
        if (Eigen_1_plus > 0 and Eigen_2_m < 0) or (Eigen_2_m > 0 and Eigen_1_plus < 0):
            elifSaddle.append(V_eq_plus)
            AlphaELIFSaddle.append(alpha)
        elif (Eigen_1_plus > Eigen_2_m) > 0 or (Eigen_2_m > Eigen_1_plus ) > 0:
            elifStable.append(V_eq_plus)
            AlphaELIFStable.append(alpha)
        elif (Eigen_1_plus and Eigen_2_m < 0) or (Eigen_2_m and Eigen_1_plus <0 ):
            elifUnstable.append(V_eq_plus)
            elifUnstable.append(alpha)


    else:
        if (Eigen_2_m.real and Eigen_1_plus.real) > 0:
            elseUnstableSpiral.append(V_eq_plus)
            AlphaELSEUnstableSpiral.append(alpha)
        elif (Eigen_2_m.real and Eigen_1_plus.real) < 0:
            elseStableSpiral.append(V_eq_plus)
            AlphaELSEStableSpiral.append(alpha)
```
