```python
ListB = np.linspace(-10, 0, num = 10000)
ListB[ListB == 0] = np.nan


alpha = -0.1
epsilon = 0.01

V_equationList = []
V_Eigen_PositiveList = []
V_Eigen_NegativeList = []

BetaIFSaddle = []
BetaFStable = []
BetaIFUnstable = []
BetaELIFSaddle = []
BetaELIFStable = []
BetaELIFUnstable = []
BetaELSEUnstableSpiral = []
BetaELSEStableSpiral = []
ifSaddle = []
ifStable = []
ifUnstable = []
elifSaddle = []
elifStable = []
elifUnstable = []
elseUnstableSpiral = []
elseStableSpiral = []

for beta in ListB:
    V_eq_plus = 0.5*((alpha+1)+(cmath.sqrt((alpha+1)**2-4*(alpha+1/beta))))
    V_equationList.append(V_eq_plus)
    T = 2*(alpha + 1) * V_eq_plus - alpha - 3*V_eq_plus**2 - epsilon * beta
    D = (2*(alpha +1 ) * V_eq_plus - alpha - 3*V_eq_plus**2 ) * (- epsilon*beta) + epsilon
    Eigen_1_plus = 0.5*(T + cmath.sqrt(T**2 - 4*D))
    Eigen_2_m =0.5*(T - cmath.sqrt(T**2 - 4*D))
    V_Eigen_PositiveList.append(Eigen_1_plus)
    V_Eigen_NegativeList.append(Eigen_2_m)


    T_sq_minus_D = T**2 - 4*D
    sqrt_T_sq_minus_D = cmath.sqrt(T ** 2 - 4 * D)

    if T_sq_minus_D.real > 0:
        if (Eigen_1_plus.real > 0 and Eigen_2_m.real < 0) or (Eigen_2_m.real > 0 and Eigen_1_plus.real < 0):
            ifSaddle.append(V_eq_plus)
            BetaIFSaddle.append(beta)
        elif (Eigen_1_plus > Eigen_2_m) > 0 or (Eigen_2_m > Eigen_1_plus ) > 0:
            ifStable.append(V_eq_plus)
            BetaFStable.append(beta)
        else:
            ifUnstable.append(V_eq_plus)
            BetaIFUnstable.append(beta)

    elif T_sq_minus_D == 0:
        if (Eigen_1_plus > 0 and Eigen_2_m < 0) or (Eigen_2_m > 0 and Eigen_1_plus < 0):
            elifSaddle.append(V_eq_plus)
            BetaELIFSaddle.append(beta)
        elif (Eigen_1_plus > Eigen_2_m) > 0 or (Eigen_2_m > Eigen_1_plus ) > 0:
            elifStable.append(V_eq_plus)
            BetaELIFStable.append(beta)
        elif (Eigen_1_plus and Eigen_2_m < 0) or (Eigen_2_m and Eigen_1_plus <0 ):
            elifUnstable.append(V_eq_plus)
            elifUnstable.append(beta)


    else:
        if (Eigen_2_m.real and Eigen_1_plus.real) > 0:
            elseUnstableSpiral.append(V_eq_plus)
            BetaELSEUnstableSpiral.append(beta)
        elif (Eigen_2_m.real and Eigen_1_plus.real) < 0:
            elseStableSpiral.append(V_eq_plus)
            BetaELSEStableSpiral.append(beta)
```
