```python
import matplotlib.pyplot as plt
import numpy as np

Beta = np.linspace(-20, 20, num=100000) # generating the beta in range -20 to 20

alpha = -0.1 # fixed alpha parameter


# Solving the V equation
V_Zero = 0 * Beta # zero equlibrum
V_positive = 0.5 * ((alpha + 1) + np.sqrt((alpha + 1) ** 2 - 4 * (-1) * (-alpha - 1 / Beta)))
V_negative = 0.5 * ((alpha + 1) - np.sqrt((alpha + 1) ** 2 - 4 * (-1) * (-alpha - 1 / Beta)))

#Plotting bifurcations
fig = plt.figure(figsize = (10, 8))
plt.title(r"Bifurcation diagram for $\beta$", fontsize=16)
plt.plot(Beta, V_Zero, color ='red')
plt.plot(Beta, V_positive, color ='blue')
plt.plot(Beta, V_negative, color ='orange')
plt.grid(linestyle='--', linewidth=0.5)
plt.grid(linestyle='--', linewidth=0.5)
plt.ylabel("V", fontsize=12)
plt.xlabel(r"$\beta$", fontsize=12)
plt.ylim(-5,5)
#plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =14, borderaxespad=6.5)
plt.savefig('bifurcation_beta_new.jpeg', dpi=300, bbox_inches="tight")
plt.show()
```
