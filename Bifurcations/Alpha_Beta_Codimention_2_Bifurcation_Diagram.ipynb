```python
#all imports
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

```python
listOfAlphas = np.linspace(-6, 6, num = 1000) # generating Alpha values
listOfBetas = np.linspace(-5, 50, num = 1000) # generating Beta values

# list for storing number of equlibrias alpha
OneEqulibriaListAlpha = []
TwoEqulibriaListAlpha = []
ThreeEqulibriasListAlpha = []

# list for storing number of equlibrias beta
OneEqulibriaListBeta = []
TwoEqulibriaListBeta = []
ThreeEqulibriasListBeta = []

# two for loops generating for every alpha, generates beta
for alpha in listOfAlphas:
    for beta in listOfBetas:
        #equation1 = (alpha + 1) ** 2
        #equation2 = 4 * (alpha + 1 / beta)
        equation1 = (alpha-1)**2
        equation2 = 4/beta

        if equation1 < equation2: # if equation 1 less than equation two
                                  # we have one equlibria
            OneEqulibriaListAlpha.append(alpha)
            OneEqulibriaListBeta.append(beta)
        elif equation1 == equation2: # if eq1 equals eq2 - we have two equlibria
            TwoEqulibriaListAlpha.append(alpha)
            TwoEqulibriaListBeta.append(beta)
        elif equation1 > equation2: # if eq1 bigger than eq2 - we have three equlibria
            ThreeEqulibriasListAlpha.append(alpha)
            ThreeEqulibriasListBeta.append(beta)
```

```python
# plotting Codimension 2 bifurcation for alpha and beta
x = np.linspace(0,100,1000)
x[ x==0 ] = np.nan

function1 = 1-2/np.sqrt(x)
function2 = 1+2/np.sqrt(x)

plt.figure(figsize=(10, 10))
plt.title(r'Codimension 2 bifurcations $\alpha$ and $\beta$', fontsize=16)
ax = plt.axes()
ax.scatter(OneEqulibriaListBeta, OneEqulibriaListAlpha, color='b', label ='One equilibria')
ax.scatter(TwoEqulibriaListBeta, TwoEqulibriaListBeta, color='white', label ='One equilibria')
ax.scatter(ThreeEqulibriasListBeta, ThreeEqulibriasListAlpha, color='r', label ='Three equilibria')

ax.plot(x, function1, color='w', linewidth=2, label = 'Two equilibria')
ax.plot(x, function2, color='w', linewidth=2)


plt.xlabel(r'$\beta$', fontsize =14)
plt.ylabel(r'$\alpha$', fontsize =14)
plt.ylim(-6,6)
plt.xlim(-5,50)
plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =14, borderaxespad=6.5)

plt.savefig('codimension_new', dpi=300, bbox_inches="tight")
```
