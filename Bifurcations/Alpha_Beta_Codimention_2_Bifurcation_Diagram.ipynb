```python
#all imports
import numpy as np
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

```python
#a = 1
#b = 1

listA = np.linspace(-6,6, num = 1000)
listB = np.linspace(-5,50, num = 1000)



OneA = []
TwoA = []
ThreeA = []
OneB = []
TwoB = []
ThreeB = []



for a in listA:
    #print(f"Alpha {a}")
    #listofAlphas.append(a)
    for b in listB:
        #listofBetas.append(b)
        #print(f"Beta {b}")
        test = (a+1)**2
        #print(f"Test {test}")
        test2 = 4*(a+1/b)
        #print(f"Test {test2}")
        if test < test2:
            OneA.append(a)
            OneB.append(b)
        elif test == test2:
            TwoA.append(a)
            TwoB.append(b)
        elif test > test2:
            ThreeA.append(a)
            ThreeB.append(b)
    #print(len(ThreeA))
    #plt.plot(listofAlphas, listofBetas)


#print(listA)
#print(TwoA)
```

```python
from matplotlib.axis import Tick
x = np.linspace(0,100,1000)
x[ x==0 ] = np.nan

function1 = 1-2/np.sqrt(x)
function2 = 1+2/np.sqrt(x)

plt.figure(figsize=(10, 10))
plt.title(r'Codimension 2 bifurcations $\alpha$ and $\beta$', fontsize=16)
ax = plt.axes()
ax.scatter(OneB, OneA, color='b', label = 'One equilibria')
ax.scatter(ThreeB, ThreeA, color='r', label = 'Three equilibria')

ax.plot(x, function1, color='w', linewidth=2, label = 'Two equilibria')
ax.plot(x, function2, color='w', linewidth=2)


plt.xlabel(r'$\beta$', fontsize =14)
plt.ylabel(r'$\alpha$', fontsize =14)
plt.ylim(-6,6)
plt.xlim(-5,50)
plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =14, borderaxespad=6.5)

plt.savefig('codimension_new', dpi=300, bbox_inches="tight")
```
