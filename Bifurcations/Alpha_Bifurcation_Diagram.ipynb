```python
import matplotlib.pyplot as plt
import numpy as np

# Creating vectors X and Y
alpha = np.linspace(-10, 10, num=100000) # generating the alpha in range -10 to 10

beta = 2 # fixed alpha parameter

# Solving the V equation
V_Zero = 0 * alpha # zero equlibrum
V_positive = 0.5 * ((alpha + 1) + np.sqrt((alpha + 1) ** 2 - 4 * (-1) * (-alpha - 1 / beta)))
V_negative = 0.5 * ((alpha + 1) - np.sqrt((alpha + 1) ** 2 - 4 * (-1) * (-alpha - 1 / beta)))

#Plotting bifurcations
fig = plt.figure(figsize = (10, 8))
plt.title(r"Bifurcation diagram for $\alpha$", fontsize=16)
plt.plot(alpha, V_negative, color ='red')
plt.plot(alpha, V_positive, color ='blue')
plt.plot(alpha, V_Zero, color ='orange')
plt.grid(linestyle='--', linewidth=0.5)
plt.grid(linestyle='--', linewidth=0.5)
plt.ylabel("V", fontsize=12)
plt.xlabel(r"$\alpha$", fontsize=12)
plt.ylim(-7.5,7.5)
plt.savefig('bifurcation_alpha_new.jpeg', dpi=300, bbox_inches="tight")
plt.show()
```
