```python
def eigenvaluesPlus(alpha, beta, V_eq, epsilon):
    return 0.5*(2 * (alpha + 1) * V_eq - alpha - 3 * V_eq ** 2 - beta) + np.sqrt((2*(alpha+1)*V_eq - alpha - 3*V_eq - beta)**2 -4*((2*(alpha + 1)* V_eq - alpha - 3*V_eq**2)*(-epsilon*beta)+ epsilon))
```

```python
def eigenvaluesMinus(alpha, beta, V_eq, epsilon):
    return 0.5*(2 * (alpha + 1) * V_eq - alpha - 3 * V_eq ** 2 - beta) - np.sqrt((2*(alpha+1)*V_eq - alpha - 3*V_eq - beta)**2 -4*((2*(alpha + 1)* V_eq - alpha - 3*V_eq**2)*(-epsilon*beta)+ epsilon))
```

```python
def eigenvaluesZeroPlus(alpha, beta, V_eq, epsilon):
    V = 0.5*((-alpha - beta * epsilon)+ cmath.sqrt((-alpha - beta*epsilon)**2 - 4 *(alpha*epsilon*beta+epsilon)))
    return V
```

```python
def eigenvaluesZeroMinus(alpha, beta, V_eq, epsilon):
    V = 0.5*((-alpha - beta * epsilon)- cmath.sqrt((-alpha - beta*epsilon)**2 - 4 *(alpha*epsilon*beta+epsilon)))
    return V
```
