```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.legend import Legend
```

```python
a1 = -0.1
e1 = 0.01
b1 = 2
```

```python
params=[b1]
ms = 500
timesteps = 1000000


def fhn(x, t, params):
    # constants

    V1 = x[0]
    W1 = x[1]

    b1 = params[0]

    dV1dt = (V1 * (a1 - V1) * (V1 - 1) - W1)
    dW1dt = (e1 * (V1 - b1 * W1))
    return [dV1dt,dW1dt]


x0 = [0.1,0] # initial conditions
t = np.linspace(0,ms,timesteps) # time
x = odeint(fhn,x0,t, args=(params,))

V1 = x[:,0]
W1 = x[:,1]
```

```python


params = [2] # alpha
x1 = odeint(fhn, x0, t,args=(params,))

params = [2.2] # alpha
x2 = odeint(fhn, x0, t,args=(params,))

params = [1.8] # alpha
x3 = odeint(fhn, x0, t,args=(params,))

f, (ax1) = plt.subplots(figsize=(10,8))
f.suptitle(r'Pertubating of the $\alpha$ parameter', fontsize=16)
line0, = ax1.plot(t, x1[:,0], color = "b", label = r"$\beta$ = 2")
line1, = ax1.plot(t, x2[:,0], color = "r", label = r"$\beta$ = 2.2")
line2, = ax1.plot(t, x3[:,0], color = "c", label = r"$\beta$ = 1.8")
ax1.set_xlabel('Time (ms)', fontsize =14)
ax1.set_ylabel('Voltage (mV)', fontsize =14)
ax1.tick_params(labelsize=13)
ax1.set_yticks([-0.5, -1, 0, 0.5, 1, 1])
ax1.grid(linestyle='--', linewidth=0.5)
ax1.grid(linestyle='--', linewidth=0.5)
leg = plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =14, borderaxespad=6.5)
f.savefig('Beta_pertubated_10_percent.jpeg', dpi=300, bbox_inches="tight")
```
