```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
```

```python
def find_max_min_index(list1, range):
    range
    MaxFromList = list1.max()
    MinFromList = list1.min()
    IndexOfMax = np.where(list1 == np.max(list1))
    IndexOfMin = np.where(list1 == np.min(list1))
    print(f"The maximum value is {MaxFromList}, with index of {IndexOfMax} and minimum value {MinFromList}, with index of {IndexOfMin} in a {range} range")
```

```python
def find_period_amplitude_and_frequency(min, max, period):
    amplitude = max + abs(min)
    period_to_seconds = period /1000
    frequency = 1/period_to_seconds
    print(f"The amplitude =  {amplitude}, the period = {period_to_seconds} and the frequency = {frequency}")
```

```python
def L2Distance(x,y):
    L2distance = np.sqrt(np.sum((x-y)**2))
    return L2distance*ms/(timesteps-1)
```

```python
a1 = -0.1
e1 = 0.01
b1 = 2
```

```python
params=[a1]
ms = 500
timesteps = 1000000


def fhn(x, t, params):
    # constants

    V1 = x[0]
    W1 = x[1]

    a1 = params[0]

    dV1dt = (V1 * (a1 - V1) * (V1 - 1) - W1)
    dW1dt = (e1 * (V1 - b1 * W1))
    return [dV1dt,dW1dt]


x0 = [0.1,0] # initial conditions
t = np.linspace(0,ms,timesteps) # time
x = odeint(fhn,x0,t, args=(params,))

V1 = x[:,0]
W1 = x[:,1]
```

```python
params = [-0.1] # alpha
x1 = odeint(fhn, x0, t,args=(params,))

params = [-0.025] #alpha
x2 = odeint(fhn, x0, t,args=(params,))

params = [-0.175] #alpha
x3 = odeint(fhn, x0, t,args=(params,))


f, (ax1) = plt.subplots(figsize=(10,8))
line0, = ax1.plot(t, x1[:,0], color = "#0000ff", label = r"$\alpha$ = -0.1")
f.suptitle(r'Pertubating of the $\alpha$ parameter', fontsize=16)

line1, = ax1.plot(t, x2[:,0], color = "#66cc00", label = r"$\alpha$ = -0.025")
line2, = ax1.plot(t, x3[:,0], color = "#ff0000", label = r"$\alpha$ = -0.175")
ax1.set_yticks([-0.5, -0.5, 0, 0.5, 1, 1])
ax1.set_xlabel('Time (ms)', fontsize =14)
ax1.set_ylabel('Voltage (mV)', fontsize =14)
ax1.tick_params(labelsize=13)
ax1.grid(linestyle='--', linewidth=0.5)
ax1.grid(linestyle='--', linewidth=0.5)
leg = plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =14, borderaxespad=6.5)
f.savefig('Alpha_pertubated_75_percent.jpeg', dpi=300, bbox_inches="tight")
```
