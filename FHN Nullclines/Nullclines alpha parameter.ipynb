```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline


"""Scaling constants k1"""
k1 = 125
"""Alpha values"""
a1 = -0.5
a2 = 0
a3 = 0.5
"""Epsilon values"""
e1 = 0.01
"""Beta values"""
b1 = 2


# setting the range
xmin = -1
xmax = 1.5
ymin = -0.5
ymax = 0.5

def fhn(x,t):
    V1 = x[0]
    R1 = x[1]

    return [k1*(V1 * (a1 - V1) * (V1 - 1) - R1),
            k1*(e1 * (V1 - b1 * R1))]



plt.figure(figsize=(10, 10))

xs=odeint(fhn, [0.5, 0.09], np.linspace(0, 100, 1000)) # integrating

plt.xlabel('V')
plt.ylabel('R')
plt.tick_params(labelsize=15)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax);
# Plot the nullclines.
x=np.arange(xmin, xmax, 0.01)
plt.plot(x, (x/b1), 'r-', label=r'$R=\dfrac{1}{\beta}V$') 
plt.plot(x, (x*(a1-x)*(x-1)) , 'b-',label='$a_1$=-0.5')
plt.plot(x, (x*(a2-x)*(x-1)) , 'c--',label='$a_1$=0')
plt.plot(x, (x*(a3-x)*(x-1)) , 'g--',label='$a_1$=0.5')
#plt.ylim(-0.1,0.2)
plt.legend()
plt.axvline(x=0,c='black') # vertical line (y axis)
plt.axhline(y=0,c='black') # horizontal line (x axis)

plt.show()
```
