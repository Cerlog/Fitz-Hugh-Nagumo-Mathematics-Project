```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import cmath

"""Alpha values"""
a2 = -0.1
"""Epsilon values"""
e1 = 0.01
"""Beta values"""
b1 = 2

# setting the range
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# solving FHN
def fhn(x,t):
    V1 = x[0]
    R1 = x[1]

    return [(V1 * (a1 - V1) * (V1 - 1) - R1),
            (e1 * (V1 - b1 * R1))]



xs=odeint(fhn, [0.1,0], np.linspace(0, 10000, 100000)) # integrating


# plotting
plt.figure(figsize=(10, 8))
plt.title(r'FHN limit cycle', fontsize=16)
plt.xlabel('V', fontsize=12)
plt.ylabel('R', fontsize=12)
#plt.tick_params(labelsize=15)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax);
# Plot the nullclines.
x=np.arange(xmin, xmax, 0.01)
plt.plot(x, (x/b1), 'b-', label=r'$R=\dfrac{1}{\beta}V$')
plt.plot(x, (x*(a2-x)*(x-1)), 'r-',label=r'$R=V\cdot(\alpha-V)\cdot(V-1)$')
plt.plot(xs[:,0], xs[:,1], "k-", label="Limit cycle",linewidth=2.5 )
plt.grid(linestyle='--', linewidth=0.5)
plt.grid(linestyle='--', linewidth=0.5)
plt.legend(bbox_to_anchor=(0.85,0.5), loc="center left", fontsize =12, borderaxespad=7.5)
plt.ylim(-0.5,0.5)
plt.xlim(-1,1.5)
plt.axvline(x=0,c='black',alpha=0.5) # vertical line (y axis)
plt.axhline(y=0,c='black',alpha=0.5) # horizontal line (x axis)
plt.savefig('FHN_limit_Cycle', dpi=300, bbox_inches="tight")
plt.show()
```
